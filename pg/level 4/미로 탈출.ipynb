```python
from collections import defaultdict
import heapq as hq

def solution(n, start, end, roads, traps):
    dic = defaultdict(lambda:float('inf'))

    Traps = [False] * (n+1)

    for t in traps:
        Traps[t] = True

    for a,b,c in roads:
        if dic[(a,b)] > c:
            dic[(a,b)] = c

    path = list(dic.keys())
    TF = [True] * len(path)
    maps = [[]for _ in range(n+1)]

    for idx,(a,b) in enumerate(path):
        maps[a].append(idx)
        maps[b].append(idx)

    ch = defaultdict(lambda:float('inf'))
    ch[(start,tuple(TF))] = 0
    que = [(0,start,TF)]
    ans = float('inf')

    while que:
        cnt,st,TF = hq.heappop(que)
        if st == end:
            return cnt
        for idx in maps[st]:

            if TF[idx] and path[idx][0] == st:
                en,cost = path[idx][1],dic[path[idx]]

            elif not TF[idx] and path[idx][1] == st:
                en,cost = path[idx][0],dic[path[idx]]

            else:
                continue

            if Traps[en]:
                tmp_TF = TF[:]
                for idx in maps[en]:
                    tmp_TF[idx] = not tmp_TF[idx]

                if ch[(en,tuple(tmp_TF))] > cnt + cost:
                    ch[(en,tuple(tmp_TF))] = cnt + cost
                    hq.heappush(que,(cnt+cost,en,tmp_TF))
            else:
                if ch[(en,tuple(TF))] > cnt + cost:
                    ch[(en,tuple(TF))] = cnt + cost
                    hq.heappush(que,(cnt+cost,en,TF))
```
